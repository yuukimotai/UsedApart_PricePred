### Rではじめる統計基礎講座　セクション４
### 相関係数 
$$\frac{共分散}{xの標準偏差 ✖️ yの標準偏差} $$
$$ r_xy = \frac{\Sigma(x_i - \bar{x})(y - \bar{y}) / n }{\sqrt{\Sigma(x_i - \bar{x})^2 /n)}\sqrt{\Sigma(y_i - \bar{y})^2 /n)}} $$
$$ = \frac{\Sigma(x_i - \bar{x})(y - \bar{y})}{\sqrt{\Sigma(x_i - \bar{x})^2}\sqrt{\Sigma(y_i - \bar{y})^2}} $$

\
■相関の強さと散布図について
 $$ 0 < r \leqq \pm{0.2} : ほとんど相関なし$$
 $$\pm{0.2} \leqq r \leqq \pm{0.4} : 弱い相関あり$$
 $$\pm{0.4} \leqq r \leqq \pm{0.7} : 相関あり$$
 $$\pm{0.7} \leqq r < \pm{1.0} : 強い相関あり$$
※データの数や外れ値によって印象がかなり変わるため、散布図を出して必ずチェックする事
また、相関関係と因果関係は必ずしも一致しないことにも注意。  
■因果関係が成立する条件  
・相関関係がある  
・時間的順序  
・第3因子が存在しない  


### 最小二乗法
\
・誤差の2乗の総和が最小になる直線を見つける
$$実際の値: y_i$$
$$直線上の値: \hat{y_i} = bx_i + a$$
$$誤差: e_i = y_i - (bx_i + a)$$
$$誤差の２乗の総和: \Sigma e_i^2$$
　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　↓
$$b=\frac{共分散}{xの分散} $$
$$a=yの平均 - b × xの平均$$

# 回帰直線に関する重要な項目  
決定係数 $ r^2 $ : 求めた回帰直線の当てはまりの良さを表す値  
→ 相関係数 r の２乗　$0 \leqq r^2 \leqq 1$ →１に近いほど当てはまりがいい

$$ yの全変動 = 回帰直線 + 残差 $$
当てはまり = 回帰直線の説明力  
## 回帰分析でも因果関係の有無に注意
$$　[原因] 説明変数 独立変数$$
　　　　　　　　　　　　　　　　　　　　　　　
            　　　　　↓　　　↑ × 原因があって結果があっても、結果から原因が確定できるわけではない                      
$$[結果]　被説明変数 従属変数$$

## 重回帰分析  
\
説明変数が2つ以上の回帰分析。
・変数が1つのときは直線 → 2つの時は平面を当てはめる  
例えば、販売個数yに関係した要素としてポスティング$x_1$,新聞広告$x_2$とした時、以下のような式が成り立つ。
$$y = a + b_1x_1 + b_2x_2$$

###　説明変数が2つの時の重回帰分析
基本的には回帰分析と同じく最小二乗法により求める。
$$\Sigma e_i^2 = {y_i - (a + b_1x_1 + b_2x_2)}^2$$
動画内の例題では、  
$$ y = 998.4 + 40.4x_1 + 9.8x_2$$

## 説明変数について
似たような説明変数のカラムは取り除く事。
例えば、マンションの価格に関するデータで、 築年数/駅からの徒歩の時間/管理会社の資本金/土地面積 という項目があった場合、  
管理会社の資本金  は関係がほぼないと思われる。そのためモデル構築などにおいてはカラムを落とす。  
また、相関係数が高く似たような項目も省く。
例題では駅までかかる時間と駅までの距離どちらかを選んでいた
